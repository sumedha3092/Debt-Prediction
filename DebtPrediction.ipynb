A.Read the dataset file “Credit.csv” (from github using the following command), and assign it to a Pandas DataFrame:

```python
# Importing the required packages and libraries
import numpy as np
import pandas as pd

# The following lines will import LinearRegression "Class",Accuracy score, and Train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

B.Check out the dataset. The “Credit” dataset includes the “balance” column (average credit card debt for a number of individuals) as target, as well as several features: age, cards (number of credit cards), education (years of education), income (in thousands of dollars), limit (credit limit), marital status, and rating (credit rating).

```python
# reading the dataset as a CSV file directly from the book website
Ad_data = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Credit.csv")

# showing the first 5 rows
Ad_data.head()
```

```python
# Creating the Feature Matrix:

feature_cols = ['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education', 'Married']
X = Ad_data[feature_cols]

X[0::10]
```

```python
# check the size:
print(X.shape)
```

```python
# selecting a Series of labels (the last column) from the DataFrame
y = Ad_data['Balance']

# checking the label vector by printing every 10 values
y[::10]
```

```python
# checking the size:
print(y.shape)
```

C.Generate the feature matrix and target vector (target is “balance” in this dataset). Then, normalize (scale) the features (note: don’t normalize the target vector!). To normalize the data, you can simply use scale(X)  from sklearn.

```python
from sklearn.preprocessing import MinMaxScaler

Ad_data_new= Ad_data

Features = MinMaxScaler()
Ad_data_new[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education', 'Married']] = Features.fit_transform(Ad_data_new[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education', 'Married']])

Ad_data_new
```

D.Split the dataset into testing and training sets with the following parameters: test_size=0.24, random_state=4.

```python
#splitting the dataset into testing and training sets with the following parameters: test_size=0.24, random_state=4.

X1_normalized = Ad_data_new[feature_cols]
y1 = Ad_data_new['Balance']

X_train, X_test, y_train, y_test = train_test_split(X1_normalized, y1, test_size=0.24, 
                                                    random_state=4)
```

E.Use Linear Regression to train a linear model on the training set. Check the coefficients of the linear regression model. Which feature is the most important? Which feature is the least important?

```python
# In the following line, "my_linreg" is instantiated as an "object" of LinearRegression "class". 

my_linreg = LinearRegression()

# fitting the model to the training data:
my_linreg.fit(X_train, y_train)

print(my_linreg.intercept_)
print(my_linreg.coef_)
```

#E Answer:
Most Important feature :Rating
Least Important feature : Income

F.Predict “balance” for the users in testing set. Then, compare the predicted balance with the actual balance by calculating and reporting the RMSE (as we saw in lab tutorial 4).

```python
# Predicting the "balance"
y_prediction = my_linreg.predict(X_test)

print(y_prediction)
```

```python
#comparing the predicted balance with the actual balance by calculating and reporting the RMSE

from sklearn import metrics

# Calculating "Mean Square Error" (MSE):
mse = metrics.mean_squared_error(y_test, y_prediction)

# Using numpy sqrt function to take the square root and calculate "Root Mean Square Error" (RMSE)
rmse = np.sqrt(mse)

print('RMSE: ',rmse)
```

G.Now, use 10-fold Cross-Validation to evaluate the performance of a linear regression in predicting the balance. Thus, rather than splitting the dataset into testing and training, use Cross-Validation to evaluate the regression performance. What is the RMSE when you use cross validation?

```python
# importing the method for 10-fold Cross-Validation:
from sklearn.model_selection import cross_val_score

feature_cols = ['Income','Limit','Rating','Cards','Age','Education','Married']
X = Ad_data_new[feature_cols] 
Y = Ad_data_new['Balance']

print(X.shape)
print(Y.shape)
```

```python
# Applying 10-fold cross validation with "linear regression":

# In the following line, "my_linregr" is instantiated as an "object" of LinearRegression "class". 
linear_reg = LinearRegression()

mse_list = cross_val_score(linear_reg, X, Y, cv=10, scoring='neg_mean_squared_error')

print(mse_list)
```

```python
# Notice that "cross_val_score" by default provides "negative" values for "mse" to clarify that mse is error.
# in order to calculate root mean square error (rmse), we have to make them positive!
mse_list_positive = -mse_list

# using numpy sqrt function to calculate rmse:
rmse_list = np.sqrt(mse_list_positive)
print(rmse_list)
```

```python
# calculate the average RMSE as final result of cross validation:
print('average_RMSE :',rmse_list.mean())
```
